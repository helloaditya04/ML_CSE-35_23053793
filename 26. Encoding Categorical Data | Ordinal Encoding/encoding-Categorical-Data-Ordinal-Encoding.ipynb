# 1. Import Libraries

```python
import pandas as pd
import numpy as np
```

# 2. Reading a local CSV File

```python
df = pd.read_csv('customer.csv')
df.head()
```

# 3. Selecting Column for train_test_split

```python
df=df.iloc[:,2:]
df.head()
```

# 4. Apply Train-Test-Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:2], df.iloc[:,-1], test_size=0.25)
```

```python
X_train.head()
```

```python
X_test.head()
```

# 5. Apply Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Poor', 'Average', 'Good'], ['School', 'UG', 'PG']])
oe=oe.fit(X_train)               # Fit the Training Dataset using Ordinal Encoding
```

```python
oe.get_params()
```

# 6. Transform the Training and Testing Dataset using Ordinal Encoding

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```

```python
X_train
```

```python
oe.categories_
```
