```python
# Similarly we can get all the required information.
c_name = []
ratings = []
rating_count = []
salary = []
intervies = []
jobs = []

company=soup.find_all('div',class_="companyCardWrapper")
for i in company:
    # Comapany Name:
    name=i.find('h2',class_="companyCardWrapper__companyName")
    if name:
        c_name.append(name.text.strip())

    # Rating:  
    rating=i.find('div', class_= "rating_text rating_text--md")
    if rating:
        ratings.append(rating.text.strip())
    else:
        ratings,append(None)

    rate_count=i.find('span', class_="companyCardWrapper__companyRatingCount")

    # Rating Count :
    action=i.find_all('span', class_="companyCardWrapper__ActionCount")
    if action[0]:
        rating_count.append(action[0].text.strip())
    else:
        rating_count.append(None)

    # Salary :
    if action[1]:
        salary.append(action[1].text.strip())
    else:
        salary.append(None)

    # interviews:
    if action[2]:
        intervies.append(action[2].text.strip())
    else:
        intervies.append(None)

    #jobs:
    if action[3]:
        jobs.append(action[3].text.strip())
    else:
        jobs.append(None)
    
c_name,ratings,rating_count,salary,intervies,jobs
```

```python
# Now we have to create a dataframe for this above result
d = {'Company_Name':c_name,
     'Ratings':ratings,
     'Rating_Count':rating_count,
     'Sakary':salary,
     'Interview':intervies,
     'Jobs':jobs    
    }
df = pd.DataFrame(d)
df.index = df.index+1
df
```

```python
# So Final Whole Data for all pages in the website
final = pd.DataFrame()
for j in range(1,75):
    url="https://www.ambitionbox.com/list-of-companies?campaign=homepage_companies_widget&page={}".format(j)
    headers={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36"}
    web_pages=requests.get(url, headers=headers).text
    soup=BeautifulSoup(web_pages, 'lxml')
    company=soup.find_all('div',class_="companyCardWrapper")

    c_name = []
    ratings = []
    rating_count = []
    salary = []
    intervies = []
    jobs = []

    company=soup.find_all('div',class_="companyCardWrapper")
    for i in company:
        # Comapany Name:
        name=i.find('h2',class_="companyCardWrapper__companyName")
        if name:
            c_name.append(name.text.strip())

        # Rating:  
        rating=i.find('div', class_= "rating_text rating_text--md")
        if rating:
            ratings.append(rating.text.strip())
        else:
            ratings,append(None)

        rate_count=i.find('span', class_="companyCardWrapper__companyRatingCount")

        # Rating Count :
        action=i.find_all('span', class_="companyCardWrapper__ActionCount")
        if action[0]:
            rating_count.append(action[0].text.strip())
        else:
            rating_count.append(None)

        # Salary :
        if action[1]:
            salary.append(action[1].text.strip())
        else:
            salary.append(None)

        # interviews:
        if action[2]:
            intervies.append(action[2].text.strip())
        else:
            intervies.append(None)

        #jobs:
        if action[3]:
            jobs.append(action[3].text.strip())
        else:
            jobs.append(None)

        # Creating dictionary
        d = {'Company_Name':c_name,
             'Ratings':ratings,
             'Rating_Count':rating_count,
             'Salary':salary,
             'Interview':intervies,
             'Jobs':jobs    
            }

        # Creating DataFrame
        df = pd.DataFrame(d)

        # Concatinating DataFrames
        final=pd.concat([final, df], ignore_index=True)

final.index = final.index+1
final.shape
```
